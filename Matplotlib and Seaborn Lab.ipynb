#### From the Fitbit data, we want to visually understand:

```python
fitbit.groupby('Months')['Steps'].median()
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
median_steps_by_month = fitbit.groupby('Months_encoded')['Steps'].median()

months = median_steps_by_month.index
median_steps = median_steps_by_month.values

plt.bar(months, median_steps)
plt.xlabel('Month')
plt.ylabel('Median Steps')
plt.title('Median Steps by Month')
plt.xticks(np.arange(13), [' ', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
median_steps_by_day = fitbit.groupby('Work_or_Weekend')['Steps'].median()

days = median_steps_by_day.index
median_daily_steps = median_steps_by_day.values

plt.bar(days, median_daily_steps)
plt.xlabel('Day')
plt.ylabel('Median Steps')
plt.title('Median Steps by Day')
plt.xticks(np.arange(2), ['Weekend', 'Week'])
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
levels_of_activity = (fitbit['Minutes Lightly Active'], fitbit['Minutes Very Active'], fitbit['Minutes Sedentary'])
activity_titles = ['Lightly Active', 'Very Active', 'Sedentary']

for activity, title in zip(levels_of_activity, activity_titles):
    plt.scatter(x=activity, y=fitbit['Steps'])
    plt.xlabel('Minutes of Activity')
    plt.ylabel('Steps')
    plt.title(title)
    plt.show()
```
